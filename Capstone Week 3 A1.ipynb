This assignment is to scrape the table of the Boroughs and Neighborhoods from the wikipedia website (https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M)

```python
#import the libraries that is necessary
import pandas as pd
import numpy as np
```

We can use the pandas library 'read_html()' to read and convert the table in the web to pandas DataFrame.

```python
dfs = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M', header=0)
df = dfs[0]
df
```

```python
#if Neighbourhood is 'Not assigned', then Neighbourhood will be the same as Borough
for n in range(len(df)):
    if df['Neighbourhood'][n] == 'Not assigned':
        df['Neighbourhood'][n] = df['Borough'][n]
```

```python
#Drop the rows of not assigned Borough
is_not_assigned = df.Borough != 'Not assigned'
df = df[is_not_assigned]
df
```

```python
#Group by the same Postcode to form a list of Neigbourhoods. This table only contain the Postcode and Neighbourhood. So we need to have another table which contain Postcode and Borough. So that we can merge them. 
dfs = df.groupby('Postcode').Neighbourhood.agg([('Neighbourhood', ', '.join)])
```

```python
#Extract the 'Postcode' and 'Borough' columns to form a new DataFrame 
dfd = df[['Postcode','Borough']]
```

```python
#Drop the duplicates of the Postcode column 
dfc = dfd.drop_duplicates(['Postcode'])
```

```python
#Change the index of the DataFrame in order to merge the two table
dfc.index = dfc.Postcode
```

```python
#Merge the two tables
dfm = pd.concat([dfs, dfc], axis=1, join='inner')
```

```python
dfm
```

```python
#Change the intex of the merged table
dfm.index = range(len(dfm))
```

```python
#Rearrange the sequence of the columns
result = dfm[['Postcode','Borough','Neighbourhood']]
```
